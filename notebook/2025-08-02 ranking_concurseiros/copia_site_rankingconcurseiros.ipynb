```python
def get_vaga_nomeado(elemento,registro):

    e_imgs = elemento.find_elements(By.XPATH, './/img')
    
    for e_img in e_imgs:
        registro[e_img.get_attribute('title')] = True

    return registro

def get_hash(texto):
    hash_obj = hashlib.sha256(texto.encode())
    hash_hex = hash_obj.hexdigest()

    return str(hash_hex)

def get_registro(elemeno_tbody):
    registro = dict.fromkeys(['hash'] + lista_head[1:-1] + ['Dentro das vagas','Nomeado'])
    
    e_tbody_cols = elemeno_tbody.find_elements(By.XPATH, './/td')
    
        # print(e_img.get_attribute('title'))
    # e_tbody_cols[0].get_attribute('outerHTML') 
    
    get_vaga_nomeado(e_tbody_cols[0],registro)
    
    registro['nome'] = e_tbody_cols[1].text
    registro['hash'] = get_hash(registro['nome'])
    
    for coluna in range(len(lista_head)-3):
        nome_coluna = lista_head[coluna+2]
        # print(nome_coluna)
        registro[nome_coluna] = e_tbody_cols[coluna+2].text

    return registro

def get_cabecario(driver):
    e_cabecario = driver.find_elements(By.XPATH, '//thead/tr/th')
    lista_head = []
    
    for nome in e_cabecario:
        lista_head.append(nome.text)
    
    return lista_head
```

# Trabalha nos dados obtidos

```python
lista_colunas = []
lista_colunas_base = ['hash','nome','Colocacao','Dentro das vagas','Nomeado','concurso','colocao_tratado','colocao_completo']
lista_df = []

for n in range(len(lista_sefaz)):
    # print(lista_sefaz[n]['radical'])
    df = pd.read_parquet(pasta_destino + lista_sefaz[n]['radical'] + '.parquet')
    lista_colunas = lista_colunas + list(df.columns)

    df['nome'] = df['nome'].str.upper()
    df['hash'] = df['nome'].apply(lambda x: get_hash(x))
    
    df['concurso'] = lista_sefaz[n]['radical']
    df['colocao_tratado'] = df['Colocacao'].apply(lambda x: (x.replace('º','')))

    df['Nomeado'] = df['Nomeado'].astype(bool)
    
    if len(df[df['colocao_tratado'] != '']) == 0:
        colocao_maxima = 0
    else:
        colocao_maxima = df[df['colocao_tratado'] != '']['colocao_tratado'].astype('int').max()
    # print(colocao_maxima)
    
    df_sem_colocao = df[df['colocao_tratado'] == ''].reset_index(drop=True).reset_index()[['hash']]
    dict_colocao = {value: int(key + colocao_maxima + 1) for key , value in df_sem_colocao.to_dict()['hash'].items()}
    
    df['colocao_completo'] = df.apply(
        lambda x: dict_colocao[x['hash']] if x['colocao_tratado'] == '' else int(x['colocao_tratado']),
        axis=1
    )
    
    df_resumido = df[lista_colunas_base]
    # print(len(df_resumido.columns))
    lista_df.append(df_resumido)

    # break
lista_colunas_unicas = list(set(lista_colunas))
# print(lista_sefaz[22])
```

```python
df = pd.concat(lista_df)
n_concursos = df['concurso'].nunique()
n_concorrentes = df['nome'].nunique()
n_concorrentes_100 = df[df['colocao_completo'] <= 100]['nome'].nunique()
print('Concursos: {}'.format(n_concursos))
print('Concorrentes: {:,}'.format(n_concorrentes).replace(",", "."))
print('Concorrentes: {:,}'.format(n_concorrentes_100).replace(",", "."))
```
