```python
import os
import pandas as pd
import datetime as dt
```

```python
def get_tempo(nome_arquivo):
    _, canal, _, dt_inicial_string, dt_final_string = nome_arquivo.split('_')
    dt_final_string = dt_final_string.split('.')[0]
    
    dt_inicial = dt.datetime.strptime(dt_inicial_string, "%Y%m%d%H%M%S")
    dt_final = dt.datetime.strptime(dt_final_string, "%Y%m%d%H%M%S")

    return (dt_final - dt_inicial).total_seconds()
```

```python
data = '2025-08-18'
canal = 'canal 3'
pasta_destino = 'C:/Gravações NVD/Nova pasta/' + canal + '/' + data
lista_arquivos = os.listdir(pasta_destino)
```

```python
lista_tempo = []

for nome_arquivo in lista_arquivos:
    # print(get_tempo(nome_arquivo))
    lista_tempo.append(get_tempo(nome_arquivo))

df = pd.DataFrame(lista_tempo)
horas = df.sum()/3600

print(horas)
```
