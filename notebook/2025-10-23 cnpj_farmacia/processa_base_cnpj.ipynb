### Funções

```python
def get_type(configuracao):

    tipos = {}
    for coluna in configuracao['colunas']:
        tipos[coluna] = 'str' 

    for coluna in configuracao['categorias']:
        tipos[coluna] = 'category' 

    return tipos

def corrige_df(tipo,df):
    
    if tipo not in correcoes.keys():
        return df
    
    campos_corrigidos = []
    
    for correcao in correcoes[tipo]:

        indices = df[ df['cnpjbase'] == correcao['cnpjbase'] ].index

        # Verifica se existem registros a serem alterados
        if len(indices) == 0:
            continue 
            
        # Se encontrar subistitui valores
        for key, value in correcao.items():
            print(key,value)
            if key != 'cnpjbase':
                df.loc[indice,key] = value
                campos_corrigidos.append(key)   
            
        # Cria lista com valores únicos    
        campos_corrigidos = list(set(campos_corrigidos))
        
        # Remove categorias não utilizadas
        for campos in campos_corrigidos:
            df[campos] = df[campos].cat.remove_unused_categories()
                
    return df
```

```python
def get_df_from_zip(tipo, arquivo_zip=None, digito=None):
    # Configuração interna
    charset = 'ISO-8859-1'
    configuracao= config[tipo]
    
    # Geta tipos
    tipos = get_type(config[tipo])
    
    # Se não informar o nome do arquivo, utilizar o da configuração
    if arquivo_zip is None:
        arquivo_zip = configuracao['arquivo_nome']
    
    # Digito
    print('Digito {}'.format(digito))
    if digito is not None:
        arquivo_zip = arquivo_zip.format(digito)
    
    # Descompacta arquivo
    zf = zipfile.ZipFile(arquivo_zip)

    # Obtem o nome do arquivo
    arquivo_txt = zf.infolist()[0].filename
    print("Nome do arquivo zipado: {}".format(arquivo_txt))
    
    cronometro = datetime.now()
    
    # Constroi data frame
    df = pd.read_csv(
        zf.open(arquivo_txt),
        sep=';',
        encoding=charset,
        header=None,
        names=configuracao['colunas'],
        dtype=tipos,
    )
        
    df = corrige_df(tipo,df)

    tempo = datetime.now() - cronometro
    print('Dataframe processado {}: {}'.format(arquivo_txt,tempo))

    return df

def converte_campos(tipo,df):
    # Configuração interna
    configuracao= config[tipo]
        
    cronometro = datetime.now()
    
    # Converte decimais  
    try: 
        for campo in configuracao['decimais']:
            df[campo] = df[campo].apply(lambda x: float(x.replace(',','.')))
    except:
        pass
        
    # Converte datas    
    try: 
        for campo in configuracao['datas']:
            # df[campo] = df[campo].apply(lambda x: None if x in ['00000000','0',np.nan]  else datetime.strptime(x,'%Y%m%d'))
            # df[campo] = df[campo].apply(lambda x: np.datetime64('NAT') if pd.isna(x) or x in ['00000000','0']  else date.fromisoformat(x))
            df[campo] = df[campo].apply(lambda x: np.nan if pd.isna(x) or x in ['00000000','0']  else date.fromisoformat(x))
    except:
        pass

#     # Converte datas
#     for campo in configuracao['datas']:
#         print(campo)
#         # df[campo] = df[campo].apply(lambda x: None if x in ['00000000','0',np.nan]  else datetime.strptime(x,'%Y%m%d'))
#         df[campo] = df[campo].apply(lambda x: 
#                                     np.nan if x in ['00000000','0',np.nan,np.datetime64('NAT')]  else date.fromisoformat(x))
    
    # Converte inteiros
    try:             
        for campo in configuracao['inteiros']:
            df[campo] = df[campo].astype('int')
    except:
        pass
    
    # Converte como categoria
    for campo in configuracao['categorias']:
        df[campo] = df[campo].astype('category')

    tempo = datetime.now() - cronometro
    print('Campos modificados {}: {}'.format(tipo,tempo))

    return df

def exporta_df(tipo,df,digito=None):
    # Configuração interna
    configuracao= config[tipo]
    if digito is None:
        parquet_nome = path_dataframes_base_cnpj + config[tipo]['df_nome'] + '.parquet'
    else:
        parquet_nome = path_dataframes_base_cnpj + tipo + '/' + config[tipo]['df_nome'] + str(digito) + '.parquet'
        
    cronometro = datetime.now()
    
    # Exporta Dataframe
    df.to_parquet(parquet_nome,compression='brotli')

    tempo = datetime.now() - cronometro
    print('Dataframe exportado {}: {}'.format(parquet_nome,tempo))
    
    return parquet_nome
```

```python
def processa_lista(dicionario,lista_tipos):
    
    for tipo in lista_tipos:
        cronometro = datetime.now()

        print('Carrega arquivo: [{}]'.format(tipo))
        df = get_df_from_zip(tipo)
        mb_puro = df.memory_usage(deep=True).sum()/1024/1024

        print('Converte campos: [{}]'.format(tipo))
        dicionario[tipo] = converte_campos(tipo,df)
        mb_convertido = dicionario[tipo].memory_usage(deep=True).sum()/1024/1024

        print('Gera Arquivo Parquet: [{}]'.format(tipo))
        parquet_nome = exporta_df(tipo,dicionario[tipo])
        mb_parquet = os.stat(parquet_nome).st_size/1024/1024

        tempo = datetime.now() - cronometro
        print('Espaço {:.1f}/{:.1f}/{:.1f} mb Tempo total: {}'.format(mb_puro,mb_convertido,mb_parquet,tempo))

        # df_unicos[tipo].info(memory_usage='deep'),
        print()
```
